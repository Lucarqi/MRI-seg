```python
import SimpleITK as sitk
from preprocess import resample_image
trains = sitk.ReadImage('datasets/train/all_image/patient6_T2.nii.gz')
labels = sitk.ReadImage('datasets/train/all_label/patient6_T2_manual.nii.gz')
print(trains.GetSpacing())
print(labels.GetSpacing())
ori_i = sitk.GetArrayFromImage(trains)
ori_m = sitk.GetArrayFromImage(labels)
image_ = resample_image(trains)
mask_ = resample_image(labels,is_label=True)
print(image_.GetSpacing())
print(mask_.GetSpacing())
image = sitk.GetArrayFromImage(image_)
mask = sitk.GetArrayFromImage(mask_)
print(ori_i.shape)
print(ori_m.shape)
print(image.shape)
print(mask.shape)
```
