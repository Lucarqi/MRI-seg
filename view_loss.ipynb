```python
# show loss 
import matplotlib.pyplot as plt
import pandas as pd

loss = pd.read_csv('output/loss.csv')
x = loss['epoch'].tolist()
# loss_G,loss_G_identity,loss_G_GAN,loss_G_cycle,loss_D
loss_G = loss['loss_G'].tolist()
loss_G_identity = loss['loss_G_identity'].tolist()
loss_G_GAN = loss['loss_G_GAN'].tolist()
loss_G_cycle = loss['loss_G_cycle'].tolist()
loss_D = loss['loss_D'].tolist()

plt.figure(figsize=(20,20))
for i in range(0,5):
    label = loss.columns.tolist()[i+1]
    loss_ = loss[label].tolist()
    axis = plt.subplot(2,3,i+1)
    axis.set_xlabel('epoch')
    axis.set_ylabel(label)
    axis.plot(x,loss_)
plt.show()
```
